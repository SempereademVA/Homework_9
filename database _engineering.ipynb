```python
from sqlalchemy import * 
from sqlalchemy.orm import * 
from sqlalchemy.ext.declarative import declarative_base
import time
import pandas as pd
```

```python
Base = declarative_base() 

class Station(Base):
    __tablename__ ='Station'
    id = Column(Integer, primary_key=True)
    station = Column(String(30))
    name = Column(String(255))
    latitude = Column(Float)
    longitude = Column(Float)
    elevation = Column(Float)

class Measurement(Base): 
     __abstract__ = True 

     id = Column(Integer, primary_key=True)
     Date = Column(DateTime)
     prcp = Column(Float)
     tobs= Column(Integer)

     _symbols = {} 

     @classmethod 
     def for_symbol(cls, symbol): 
         if symbol in cls._symbols: 
             return cls._symbols[symbol] 

         cls._symbols[symbol] = measurement_cls = type( 
             "%s_Station" % symbol, 
             (cls, ), 
             { 
                 "__tablename__": "measurement_%s" % symbol, 
                 "symbol": symbol 
             } 
         ) 

         return measurement_cls 
```

```python
station_df = pd.read_csv('clean_hawaii_stations.csv')
symbols = []
for index, row in station_df.iterrows():
    symbols.append(row['station'])
    
symbols
```

```python
 
#timestr = time.strftime("%Y%m%d%H%M%S")
#file_name = 'sqlite:///test' + timestr + '.sqlite'
file_name = 'sqlite:///hawaii.sqlite'
engine = create_engine(file_name, echo=True)

for symbol in symbols: 
     Measurement.for_symbol(symbol) 

Base.metadata.create_all(engine) 


sess = sessionmaker(engine)() 
```
